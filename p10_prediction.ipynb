3) According to the problem description, the measurement noise in the head observations was 0.022m, and 10% for the flux observations. Using error-based weighting, this resulted in high weights for the heads, necessitating a multiplier of 50 for the flux weights to balance the objective function. Correct (error-based) weighting implies that the "calibrated" objective function should be approximately equal to the number of observations (e.g., Doherty, 2010b, p93). Compute a factor to scale weights so that they are approximate in magnitude to measurement error:  

$$wf = \sqrt{\frac{n\:observations}{\Phi}}$$

```python
wf = np.sqrt(38 / 8.20247E+06)
wf
```

4) Compute a jacobian matrix with the 'calibrated' parameter set by running PEST with NOPTMAX in the prediction PEST control file (pred.pst) set to -1  
  
5) Prepare responses files for GENLINPRED so that prediction uncertainty calculations can be batched for all of the head measurement locations. General form of the response file:  


    <enter at first prompt>
    a # abbreviated or full calculation
    pred.pst # pest control file with predictions
    u # use parameter uncertainty file
    parameters.unc # name of parameter uncertainty file
    n # weights are not the inverse of measurement uncertainty
    0.0021523826022590139 # wf from above
    P.out # genlinpred output file
    y # perform comprehensive analysis of a prediction
    P # name of prediction observation in pest control file
    pred.jco # name of jacobian (.jco) file computed in step 4 above
    y # compute predictive uncertainty
    y # compute parameter contributions to uncertainty
    i # compute individual parameter contributions to uncertainty
    n # do not compute new data worth

    

```python
#function to make response file for a given parameter:
def make_response(pred):
    rsp = ['',
    'a',
    'pred.pst',
    'u',
    'parameters.unc',
    'n',
    '0.0021523826022590139',
    '{}.out'.format(pred),
    'y',
    pred,
    'pred.jco',
    'y',
    'y',
    'i',
    'n']
    
    ofp = open('{}.rsp'.format(pred), 'w')
    [ofp.write(l + '\r\n') for l in rsp] # note '\r\n' enforces writing of Windows line endings
    ofp.close()
```

```python
predictions = ['P_T_1yr',
'G_T_1yr',
'F_T_1yr',
'N_T_1yr',
'J_T_1yr',
'E_T_1yr',
'A_T_1yr',
'B_T_1yr',
'K_T_1yr',
'Q_T_1yr',
'M_T_1yr',
'I_T_1yr',
'D_T_1yr',
'C_T_1yr',
'O_T_1yr',
'H_T_1yr',
'S_T_1yr']
```

####Make response files for each prediction, then run genlinpred with each response

```python
import os

os.chdir('pred_unc')
```

```python
for p in predictions:
    make_response(p)
    os.system('wine genlinpred.exe <{}.rsp'.format(p))
```

####Get the total, pre- and post-calibration uncertainty standard deviation for each prediction:
(which are on lines 44 and 48 of the genlinpred output file)

```python
uncertainties = {}
for p in predictions:
    glpout = open('{}.out'.format(p)).readlines()
    unc_pre = float(glpout[43].strip().split()[5])
    unc_post = float(glpout[47].strip().split()[5])
    uncertainties[p[0].lower()] = [unc_pre, unc_post] # use the first character of prediction name as index
    
    unc = pd.DataFrame.from_dict(uncertainties, orient='index')
    unc.columns = ['pre-cal', 'post-cal']
```

```python
import os
import shutil

try:
    os.chdir('pred_unc')
    os.remove('record.dat')
except:
    pass
                 
```

####Adjust parameters (as in "adjust_parameters.bat")

```python
record = []
for i in xrange(100):
    print 'ns_random{}.par'.format(i+1)
    os.system('wine parrep ns_random{}.par pred.pst.copy pred.pst'.format(i+1))
    os.system('wine pest pred_svda.pst')
    rec = open('pred_svda.rec', 'r')
    for line in rec:
        if 'Sum of squared weighted residuals (ie phi)' in line:
            record.append(line.strip().split('=')[-1])
    shutil.copy('pred.bpa', 'pred.bpa.{}'.format(i+1))
                
# write out the records
ofp = open('record.dat','w')
[ofp.write(r + '\n') for r in record]
ofp.close()
```

####Post calibration runs (as in postcal_runs.bat)

```python
import pandas as pd

# get the prediction names from the ins file
predictions = [l.split('!')[1].lower() for l in open('resultsT_txt.ins', 'r').readlines()[1:18]]
results = pd.DataFrame(index=predictions)

for i in xrange(100):
    print 'pred.bpa.{}'.format(i+1)
    os.system('wine parrep pred.bpa.{} pred_NM0.pst temp.pst'.format(i+1))
    os.system('wine pest temp')
    
    # get the results
    r = [float(r.strip()) for r in open('resultsT.txt', 'r').readlines()[34:51]]
    results[str(i+1)] = r

# save down results to a csv file
results.to_csv('results.csv', index_label='prediction')
```
