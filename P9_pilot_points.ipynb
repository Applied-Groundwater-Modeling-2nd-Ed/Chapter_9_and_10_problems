####functions####

```python
def readk(infile):
    '''
    read k-arrays that have been wrapped by fac2real
    '''
    k=[]
    indat = open(infile,'r').readlines()
    for line in indat:
        tmp = line.strip().split()
        k.extend(tmp)
    k = np.array(k).astype(float)
    return k

def plotk(infile, extent, log=False, clim=None, points=False, wells=False):
    
    k = readk(infile)
    k = np.reshape(k, (15, 15))
    
    if log:
        k = np.log10(k)
        label = 'log k, m/d'
    else:
        label = 'k, m/d'
        
    if not clim:
        clim = np.min(k), np.max(k)

    plt.imshow(k, vmin=clim[0], vmax=clim[1], extent=extent, interpolation='None')
    cb = plt.colorbar()
    cb.set_label(label)
    
    # plot pilot points on top of k-field
    if points:
        pnts = pd.read_csv('pilot_points/pp.csv')
        plt.scatter(pnts.X, pnts.Y, c='w')
        
    # overlay observation wells on k-field    
    if wells:
        pnts = pd.read_csv('observations.csv')
        plt.scatter(pnts.X, pnts.Y, c='w')
    print 'max K: {}\nmin K: {}'.format(np.max(k), np.min(k))
```
